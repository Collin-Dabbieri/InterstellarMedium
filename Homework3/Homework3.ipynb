```python
import numpy as np
from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 15, 8
```

```python
k=8.617e-5 #eV/K
T=10000
#i<j
#correct is boolean for if there should be fine structure correction of Upsilon value
def gimmeqij(Ups,T,wi,Ei,Ej,correct,Si,Li,Ji):
    exp=-(Ej-Ei)/(k*T)
    if correct:
        Ups=Ups*(2*Ji+1)/((2*Si+1)*(2*Li+1))
    return((8.629e-6*Ups*np.exp(exp))/(wi*T**0.5))
    
def gimmeqji(Ups,T,wj,correct,Si,Li,Ji):
    if correct:
        Ups=Ups*(2*Ji+1)/((2*Si+1)*(2*Li+1))
    
    return((8.629e-6*Ups)/(wj*T**0.5))  
```

```python
## Let's tidy it all up into one nice function

#this function returns N_i/N for a 5 level atom

#E_vector is [E1,E2,E3,E4,E5]
# U_vector is [Ups(1,2),Ups(1,3),Ups(1,4),Ups(1,5),Ups(2,3),Ups(4,5)]
# w_vector is [w1,w2,w3,w4,w5]
# A_vector is [A21,A31,A41,A51,A32,A42,A52,A43,A53,A54]

# We'll set N=1 so Ni is the fraction

E_vector=[0.0,0.0140323,0.0379607,2.513565,5.354349]
w_vector=[1,3,5,5,1]
U_vector=[0.55,0.27,2.29,0.29,1.29,0.58]
A_vector=[2.61e-05,3.17e-11,2.41e-06,0,7.34e-12,4.57e-06,2.15e-01,3.52e-05,6.34e-04,1.71e00]

T=10000 #K
k=8.617e-5 #eV/K


def calculate_level_populations(T,Ne,E_vector,U_vector,A_vector,w_vector):
    #gimmeqij(Ups,T,wi,Ei,Ej)
    q12=gimmeqij(U_vector[0],T,w_vector[0],E_vector[0],E_vector[1],False,0,0,0)
    q13=gimmeqij(U_vector[1],T,w_vector[0],E_vector[0],E_vector[2],False,0,0,0)
    q14=gimmeqij(U_vector[2],T,w_vector[0],E_vector[0],E_vector[3],True,1,2,0) #Correct Upsilon for fine state
    q15=gimmeqij(U_vector[3],T,w_vector[0],E_vector[0],E_vector[4],True,1,2,0) #Correct Upsilon for fine state
    q23=gimmeqij(U_vector[4],T,w_vector[1],E_vector[1],E_vector[2],False,0,0,0)
    q24=gimmeqij(U_vector[2],T,w_vector[1],E_vector[1],E_vector[3],True,1,2,1) #Correct Upsilon for fine state
    q25=gimmeqij(U_vector[3],T,w_vector[1],E_vector[1],E_vector[4],True,1,2,1) #Correct Upsilon for fine state
    q34=gimmeqij(U_vector[2],T,w_vector[2],E_vector[2],E_vector[3],True,1,2,2) #Correct Upsilon for fine state
    q35=gimmeqij(U_vector[3],T,w_vector[2],E_vector[2],E_vector[4],True,1,2,2) #Correct Upsilon for fine state
    q45=gimmeqij(U_vector[5],T,w_vector[3],E_vector[3],E_vector[4],False,0,0,0)
    
    #gimmeqji(Ups,T,wj)
    q54=gimmeqji(U_vector[5],T,w_vector[4],False,0,0,0)
    q53=gimmeqji(U_vector[3],T,w_vector[4],True,1,2,2)#Correct Upsilon for fine state
    q52=gimmeqji(U_vector[3],T,w_vector[4],True,1,2,1)#Correct Upsilon for fine state
    q51=gimmeqji(U_vector[3],T,w_vector[4],True,1,2,0)#Correct Upsilon for fine state
    q43=gimmeqji(U_vector[2],T,w_vector[3],True,1,2,2)#Correct Upsilon for fine state
    q42=gimmeqji(U_vector[2],T,w_vector[3],True,1,2,1)#Correct Upsilon for fine state
    q41=gimmeqji(U_vector[2],T,w_vector[3],True,1,2,0)#Correct Upsilon for fine state
    q32=gimmeqji(U_vector[4],T,w_vector[2],False,0,0,0)
    q31=gimmeqji(U_vector[1],T,w_vector[2],False,0,0,0)
    q21=gimmeqji(U_vector[0],T,w_vector[1],False,0,0,0)
    
    A21=A_vector[0]
    A31=A_vector[1]
    A41=A_vector[2]
    A51=A_vector[3]
    A32=A_vector[4]
    A42=A_vector[5]
    A52=A_vector[6]
    A43=A_vector[7]
    A53=A_vector[8]
    A54=A_vector[9]
    
    X=np.zeros((5,5))
    
    X[0,0]=Ne*(q12+q13+q14+q15)
    X[1,0]=-Ne*q12
    X[2,0]=-Ne*q13
    X[3,0]=-Ne*q14
    X[4,0]=1.
    
    X[0,1]=-Ne*q21-A21
    X[1,1]=Ne*(q21+q23+q24+q25)+A21
    X[2,1]=-Ne*q23
    X[3,1]=-Ne*q24
    X[4,1]=1.
    
    X[0,2]=-Ne*q31-A31
    X[1,2]=-Ne*q32-A32
    X[2,2]=Ne*(q31+q32+q34+q35)+A31+A32
    X[3,2]=-Ne*q34
    X[4,2]=1.
    
    X[0,3]=-Ne*q41-A41
    X[1,3]=-Ne*q42-A42
    X[2,3]=-Ne*q43-A43
    X[3,3]=Ne*(q41+q42+q43+q45)+A41+A42+A43
    X[4,3]=1.
    
    X[0,4]=-Ne*q51-A51
    X[1,4]=-Ne*q52-A52
    X[2,4]=-Ne*q53-A53
    X[3,4]=-Ne*q54-A54
    X[4,4]=1.
    
    
    #X*x=b, where x=[N1,N2,N3,N4,N5], b=[0,0,0,0,N], where N==1
    
    b=np.asarray([0,0,0,0,1])
    
    x=np.linalg.solve(X,b)
    
    return(x)
    
```

```python
# Now iterate over a range of Ne values

Ne_vals=np.logspace(-5,10,num=10000)

N1=[]
N2=[]
N3=[]
N4=[]
N5=[]
for i in Ne_vals:
    x=calculate_level_populations(T,i,E_vector,U_vector,A_vector,w_vector)
    N1.append(x[0])
    N2.append(x[1])
    N3.append(x[2])
    N4.append(x[3])
    N5.append(x[4])
```

# Now we'll calculate the level populations assuming LTE
# Boltzmann equation

$\frac{N_2}{N_1}=\frac{\omega_2}{\omega_1}e^{-\frac{E}{kT}}$


```python
def Level_Populations_Boltzmann(E_vector,T,w_vector):
    
    E21=E_vector[1]-E_vector[0]
    E31=E_vector[2]-E_vector[0]
    E41=E_vector[3]-E_vector[0]
    E51=E_vector[4]-E_vector[0]
    
    w1=w_vector[0]
    w2=w_vector[1]
    w3=w_vector[2]
    w4=w_vector[3]
    w5=w_vector[4]
    
    N2N1=np.exp(-E21/(k*T))*(w2)/(w1)
    N3N1=np.exp(-E31/(k*T))*(w3)/(w1)
    N4N1=np.exp(-E41/(k*T))*(w4)/(w1)
    N5N1=np.exp(-E51/(k*T))*(w5)/(w1)
    
    #Using some algebra and the fact that N1+N2+N3+N4+N5=1
    #N1+N2N1*N1+N3N1*N1+N4N1*N1+N5N1*N1=1
    #N1(1+N2N1+N3N1+N4N1+N5N1)=1
    #N1=1/(1+N2N1+N3N1+N4N1+N5N1)
    
    N1=1/(1+N2N1+N3N1+N4N1+N5N1)
    N2=N1*N2N1
    N3=N1*N3N1
    N4=N1*N4N1
    N5=N1*N5N1

    
    return([N1,N2,N3,N4,N5])
    
    
levels=Level_Populations_Boltzmann(E_vector,T,w_vector)
levels
```

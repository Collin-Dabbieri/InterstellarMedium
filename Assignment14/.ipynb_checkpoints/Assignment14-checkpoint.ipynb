## This line appears to be H-alpha WL 6564 A

$$Q=\frac{L(H\alpha)}{h\nu}\frac{\alpha_B(H^0,T)}{\alpha^{eff}_{H\alpha}(H^0,T)}$$

$$\alpha_B=2.59x10^{-13} cm^3 s^{-1}$$

$$\alpha^{eff}_{H\alpha}=3.03x10^{-14} * 2.847 cm^3 s^{-1}=8.63x10^{-14}cm^3s^{-1}$$

$$\nu=\frac{3x10^8}{6564x10^{-10}}=4.57x10^{14} s^{-1}$$



## Now we need to find L

```python
ui.load_data(1,'j1730.dat',3)
ui.ignore_id(1,"0:6400")
ui.ignore_id(1,"8000:20000")
ui.set_source(1,ui.powlaw1d.pl1)
ui.set_source(1,ui.emissiongaussian.eg1)
ui.set_model(1,pl1+eg1)
pl1.gamma=1.8
pl1.ref=7500
pl1.ampl=12

eg1.fwhm=5000
eg1.pos=6564
eg1.flux=5000
eg1.skew=1
ui.thaw(eg1.pos)

ui.plot_fit(1)
```

```python
ui.fit()
```

```python
ui.plot_fit()
```

```python
#subtract out power law

wavelength=ui.get_data_plot(1).x
flux=ui.get_data_plot(1).y
model=ui.get_model_plot(1).y

def powlaw(x,A,ref,gamma):
    return(A*(x/ref)**(-1*gamma))



A=11.28
ref=7500
gamma=1.76032

em=[]
for i in range(len(wavelength)):
    wl=wavelength[i]
    pl=powlaw(wl,A,ref,gamma)
    #print(pl)
    em.append(model[i]-pl)
```

```python
plt.plot(wavelength,em)
```

```python
omega=0.2

flux=simps(em,wavelength)/omega #units of 10^-17 erg s^-1 cm^-2
flux=flux*1e-17
flux=flux/1e7 #units J s^-1 cm^-2 
R=15498.1*3.086e+24 #cm


Lum=4.*np.pi*flux*R**2 #J s^-1
h=6.626e-34 #J s
nu=4.57e14 #s^-1

Q=(Lum/(h*nu))*((2.59e-13)/(8.63e-14)) #s^-1

logQ=np.log10(Q)
logQ
```
